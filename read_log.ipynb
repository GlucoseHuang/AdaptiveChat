```python
import json
import pandas as pd
from pathlib import Path

# 1. 定义文件夹路径
log_dir = Path("/Users/glucose/Desktop/DialogSystem/experiment_logs/")

all_records = []

# 2. 遍历该目录下所有的 .jsonl 文件
# 如果包含子文件夹，可以使用 log_dir.rglob("*.jsonl")
for file_path in log_dir.glob("*.jsonl"):
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            if line.strip():
                record = json.loads(line)
                
                # 【可选】记录数据来源的文件名，方便后续分析是哪次实验的数据
                record['source_file'] = file_path.name 
                
                all_records.append(record)

# 3. 一次性将所有记录转换为 DataFrame
df = pd.DataFrame(all_records)

print(f"共读取了 {len(list(log_dir.glob('*.jsonl')))} 个文件")
print(f"DataFrame 总行数: {len(df)}")
df
```

```python
df.to_excel("Data.xlsx", index=False)
```
